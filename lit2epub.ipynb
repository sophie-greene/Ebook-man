```python
def verifye(f):
    m,ff=os.path.split(f)
    try:
        if os.path.exists(f):  
            book=epub.read_epub(f)
            os.remove(f)
            epub.write_epub(f,book)
            return (f,"OK")
        else:
            return (f,"DOESNT EXIST")
            #print "file  ..." ,ff," ...... verified"
    except:
        #print "file corrupt ..." ,ff," ...... deleting"
        #os.remove(f)
        return (f,"NOT OK")
```
