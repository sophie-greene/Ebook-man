```python
d="/media/some/sophie-hhd/books/thrill/"
files=!find $d -type f -iname "*.epub" #"*.mobi"  -o -iname "*.chm" -o -iname "*.lit" -o -iname "*.azw3"
len(files)
```

```python
def getMeta(f):
    m=!exiftool "$f"
    meta={}
    for mm in m:
        tmp=mm.split(" : ")
        if len(tmp)==2:
            meta[tmp[0].strip()] = tmp[1].strip()
    fp,fn=os.path.split(f)
    n, ext = os.path.splitext(fn)
    return [n,meta]
```

```python
t0 = time.time()
p = Pool(20)
metalist=p.map(getMeta, files)
p.terminate()
p.close()
t1 = time.time()
total = t1-t0
print('time: ',total)
```

```python
metaDict={}
for mm in metalist:
        metaDict[mm[0]] = mm[1]   
```
